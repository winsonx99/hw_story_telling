```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract relevant columns
REGION = df['REGION']
TEACHERS = df['TEACHERS']
MINOR_POPULATION = df['MINOR POPULATION']

# Bar plot
plt.figure(figsize=(8, 6))
plt.bar(REGION, TEACHERS)
plt.xlabel('Region')
plt.ylabel('Number of Teachers')
plt.title('Number of Teachers by Region')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(MINOR_POPULATION, TEACHERS)
plt.xlabel('Minor Population')
plt.ylabel('Number of Teachers')
plt.title('Number of Teachers vs. Minor Population')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read data from CSV file
data = pd.read_csv('044.csv')

# Grouped Bar Plot
states = data['STATE']
teachers = data['TEACHERS']
minor_population = data['MINOR POPULATION']

# Set the width of the bars
bar_width = 0.35

# Set the positions of the bars on the x-axis
x = np.arange(len(states))

# Create grouped bar plot
fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(x - bar_width/2, teachers, bar_width, label='Teachers')
bar2 = ax.bar(x + bar_width/2, minor_population, bar_width, label='Minor Population')

# Set x-axis labels and tick labels
ax.set_xticks(x)
ax.set_xticklabels(states, rotation=90)

# Set labels and title
ax.set_xlabel('State')
ax.set_ylabel('Count')
ax.set_title('Number of Teachers and Minor Population by State')

# Add legend
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('044.csv')

# Select relevant columns for correlation analysis
columns_of_interest = ['MINOR POPULATION', 'TEACHERS']
subset_data = data[columns_of_interest]

# Compute correlation matrix
correlation_matrix = subset_data.corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv('044.csv')

# Extract relevant columns
minor_population = data['MINOR POPULATION']
teachers = data['TEACHERS']

# Create subplots for histograms
fig, axs = plt.subplots(2, 1, figsize=(8, 8))

# Histogram for MINOR POPULATION
axs[0].hist(minor_population, bins=10, alpha=0.7)
axs[0].set_xlabel('Minor Population')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Histogram of Minor Population')

# Histogram for TEACHERS
axs[1].hist(teachers, bins=10, alpha=0.7)
axs[1].set_xlabel('Number of Teachers')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Histogram of Number of Teachers')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
